# Next Sentence Prediction is the second part of the training the LLM.
# NSP helps to predict if the 2 text sentences are in a sequence or not. ie, say we have 2 texts txt1 and txt2, NSP will say if txt2 is the next sentence of txt1 or not

```python
from transformers import BertTokenizer, BertForNextSentencePrediction
import torch
```

```python
text = ("After Abraham Lincoln won the November 1860 presidential election on an "
        "anti-slavery platform, an initial seven slave states declared their "
        "secession from the country to form the Confederacy.")
text2 = ("War broke out in April 1861 when secessionist forces attacked Fort "
         "Sumter in South Carolina, just over a month after Lincoln's "
         "inauguration.")
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForNextSentencePrediction.from_pretrained('bert-base-uncased')
```

```python
inputs = tokenizer(text, text2, return_tensors="pt") # both the texts are in same sentence , but they are seperated by input_id: 102
inputs.keys()
```

```python
# we need to create label tensor that can be used for training
# 0 : is next sentence
# 1 : is not next sentence
labels = torch.LongTensor([0]) # here in this case both sentences are next to each other.
labels
```

```python
outputs = model(**inputs, labels= labels)
outputs.keys()
```

```python
outputs.logits
```

```python
torch.argmax(outputs.logits)
```

```python
# we got 0 => we got the correct result saying both the texts are next to each other

outputs.loss
```
